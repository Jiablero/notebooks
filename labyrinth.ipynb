```python
import random

lab_file = '/content/Карточки по математике  - брахиозавр.csv'
result_file = 'брахиозавр.csv'
matrix = []
result = []
maxval = 0

with open(lab_file) as lab:
  lines = lab.read().split('\n')
  for l in lines[:-1]: matrix.append(l.split(','))

for m in matrix:
  for v in m:
    if int(v) > maxval: maxval = int(v)

allvals = {}
for m in matrix:
  for v in m: 
    if v not in allvals:
      val = int(v)
      if val == 0:
        sums = [f'{i} - {i}' for i in range(0, maxval)]
      else:
        sums = [f'{i} + {val - i}' for i in range(0, val)]
        if val < maxval:
          for i in range(0, 10):
            sums.append(f'{val + i} - {i}')
      random.shuffle(sums)
      allvals[v] = {'sums': sums, 'pos': 0}

for m in matrix:
  newline = []
  for l in m:
    if allvals[l]['pos'] >= len(allvals[l]['sums']) - 1: allvals[l]['pos'] = 0
    newline.append(allvals[l]['sums'][allvals[l]['pos']])  
    #print ('{} = {}'.format(l, allvals[l]['sums'][allvals[l]['pos']]))
    allvals[l]['pos'] += 1
  result.append(newline)

  with open(result_file, 'w') as fl:
    fl.write('\n'.join(','.join(j for j in i) for i in result))
```
