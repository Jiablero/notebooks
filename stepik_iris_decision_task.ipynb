```
from sklearn import tree
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```
iris = pd.read_csv('train_iris.csv')
```

```
iris.head()
```

```
X = iris.drop(['Unnamed: 0', 'species'], axis = 1)
y = iris.species
```

```
test_iris = pd.read_csv('test_iris.csv')
test_X = test_iris.drop(['Unnamed: 0', 'species'], axis = 1)
test_y = test_iris.species
```

```
np.random.seed(0)

results = pd.DataFrame()
for i in range(0, 100):
  clf = tree.DecisionTreeClassifier(criterion='entropy')
  clf.fit(X, y)
  train_score = clf.score(X, y)
  test_score = clf.score(test_X, test_y)
  r = pd.DataFrame({'depth': [i], 'train_score': [train_score], 'test_score': [test_score]})
  results = results.append(r)
```

```
results.head(20)
```

```
results = pd.melt(results, id_vars=['depth'], value_vars=['train_score', 'test_score'],
                          var_name='set_type', value_name='score')
```

```
sns.lineplot(x="depth", y="score", hue="set_type", data=results)
```
